```python
# Delete only fifa1['Release'] nan values 
fifa1= fifa.dropna(subset=["Release Clause"])   
```

```python
for i in fifa1['Release']:
    if i.endswith('M'):
        fifa1['Release']=fifa1['Release'].apply(lambda x : x.split("M")[0])
    if i.endswith('K'):
        fifa1['Release']=fifa1['Release'].apply(lambda x : x.split("K")[0]) 
    
fifa1['Release']
fifa1 
```

```python
# Barcelona best player in dribbling
barca_pl=fifa[fifa['Club']=='FC Barcelona']
barcapl=barca_pl[barca_pl['Dribbling']>=75]
barca_pl=barca_pl.loc[0:,['Name','Dribbling']]
T_barca_pl=barca_pl.sort_values(by='Dribbling',ascending=False)[0:10]
T_barca_pl
```

```python
name_=T_barca_pl['Name']
drib=T_barca_pl['Dribbling']
plt.figure(figsize=(12,5))
plt.bar(name_,drib,color=['red','gold','black','gold','black','gold','black','gold','black','gold'])
plt.show()
```
